Rank
B) Name
C) Artist
D) Upload date
E) Views 

```python
driver=webdriver.Chrome()
```

```python
driver.get(' https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos ')
```

```python
driver.maximize_window()
```

```python
rank=[]
name=[]
artist=[]
upload_date=[]
views=[]
```

```python

try:
    video_name=driver.find_elements(By.XPATH,'//table[@class="sortable wikitable sticky-header static-row-numbers sort-under col3center col4right jquery-tablesorter"]/tbody/tr/td[1]')
    for i in video_name:
        name.append(i.text)
except:
    name.append('-')
    
    
try:
    name_of_artist=driver.find_elements(By.XPATH,'//table[@class="sortable wikitable sticky-header static-row-numbers sort-under col3center col4right jquery-tablesorter"]/tbody/tr/td[2]')
    for i in name_of_artist:
        artist.append(i.text)
except:
    artist.append('-')
    
try:
    date=driver.find_elements(By.XPATH,'//table[@class="sortable wikitable sticky-header static-row-numbers sort-under col3center col4right jquery-tablesorter"]/tbody/tr/td[4]')
    for i in date:
        upload_date.append(i.text)
except:
    upload_date.append('-')
    
try:
    view=driver.find_elements(By.XPATH,'//table[@class="sortable wikitable sticky-header static-row-numbers sort-under col3center col4right jquery-tablesorter"]/tbody/tr/td[3]')
    for i in view:
        views.append(i.text)
except:
    views.append('-')
    
rank = [i for i in range(1, len(name) + 1)]    
```

```python
import pandas as pd
data = pd.DataFrame({
    'Rank': rank,
    'Name': name,
    'Artist': artist,
    'Upload Date': upload_date,
    'Views (Billions)': views
})

data
    
```

# Question 8:-

Details of Datasets from UCI machine learning repositories.
Url = https://archive.ics.uci.edu/ You
have to find the following details:
A) Dataset name
B) Data type
C) Task
D) Attribute type
E) No of instances
F) No of attribute G) Year
 Note: - from the home page you have to go to the Show All Dataset page through code. 

```python
driver=webdriver.Chrome()
driver.get('https://archive.ics.uci.edu/')
driver.maximize_window()
```

```python
try:
    click=driver.find_element(By.XPATH,'/html/body/div/div[1]/div[1]/main/div/div[1]/div/div/div/a[1]')
    click.click()
    
except:
    print('no cookies available')
```

```python
try:
    time.sleep(3)
    more=driver.find_element(By.XPATH,'/html/body/div/div[1]/div[1]/main/div/div[2]/div[1]/div/label[2]')
    more.click()
    
except:
    print('no cookies available')
```

```python
pages=[]
start=0
end=66

for page in range(start,end):
    page=driver.find_elements(By.XPATH,'//a[@class="link-hover link text-xl font-semibold"]')
    for i in page:
        pages.append(i.get_attribute('href'))
    next_page=driver.find_element(By.XPATH,'//button[@class="btn-primary btn-sm btn"][2]')
    next_page.click()
    time.sleep(2)
len(pages)
```

```python
Dataset =[]
Data_Type =[]
Task =[]
Attribute_Type =[]
No_of_Instances =[]
No_of_Attribute =[]
Year =[]

for i in pages:
    driver.get(i)
    time.sleep(2)

    try:
        dataset=driver.find_element(By.XPATH,'//h1[@class="text-3xl font-semibold text-primary-content"]')
        Dataset.append(dataset.text)
    except:
        Dataset.append('-')
        
    
    try:
        datatype=driver.find_element(By.XPATH,'//div[@class="grid grid-cols-8 gap-4 md:grid-cols-12"]/div[1]/p')
        Data_Type.append(datatype.text)
    except:
        Data_Type.append('-')
        

    try:
        task=driver.find_element(By.XPATH,'//div[@class="grid grid-cols-8 gap-4 md:grid-cols-12"]/div[3]/p')
        Task.append(task.text)
    except:
        Task.append('-')
        
        
    try:
        attributetype=driver.find_element(By.XPATH,'//div[@class="grid grid-cols-8 gap-4 md:grid-cols-12"]/div[4]/p')
        Attribute_Type.append(attributetype.text)
    except:
        Attribute_Type.append('-')
        

        
    try:
        no_of_instances=driver.find_element(By.XPATH,'//div[@class="grid grid-cols-8 gap-4 md:grid-cols-12"]/div[5]/p')
        No_of_Instances.append(no_of_instances.text)
    except:
        No_of_Instances.append('-')
        
        
    try:
        no_of_attribute=driver.find_element(By.XPATH,'//div[@class="grid grid-cols-8 gap-4 md:grid-cols-12"]/div[6]/p')
        No_of_Attribute.append(no_of_attribute.text)
    except:
        No_of_Attribute.append('-')
        
        

    try:
        year=driver.find_element(By.XPATH,'//h2[@class="text-sm text-primary-content"]')
        Year.append(year.text.split()[2])
    except:
        Year.append('-')
        
        
        
        
len(Dataset),len(Data_Type),len(Task),len(Attribute_Type),len(No_of_Instances),len(No_of_Attribute),len(Year)
```

```python
df=pd.DataFrame({'DataSet Title':Dataset,'Data Type':Data_Type,'Task':Task,'Attribute Type':Attribute_Type,
                'No of Instances':No_of_Instances,
                'No of Attribute':No_of_Attribute,
                'Year':Year})
df
```
